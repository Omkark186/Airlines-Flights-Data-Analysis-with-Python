The goal of this project is to extract insights that can benefit stakeholders in the airlines and travel industries, such as:
Identifying factors affecting flight ticket prices
Understanding the impact of departure time on ticket cost
Analyzing travel duration trends

Comparing airline performance and pricing
The Flights Booking Dataset of various Airlines is a scraped datewise from a famous website in a structured format.
The dataset contains the records of flight travel details between the cities in India. 
Here, multiple features are present like Source & Destination City, Arrival & Departure Time, Duration & Price of the flight etc.

The dataset contains detailed flight travel records with the following key features:

Airline – Name of the airline operating the flight
Source City – Departure city
Destination City – Arrival city
Departure Time – Scheduled take-off time
Arrival Time – Scheduled landing time
Duration – Total travel time
Price – Ticket price of the flight
Additional Info – Other details like layovers, travel class, etc.
